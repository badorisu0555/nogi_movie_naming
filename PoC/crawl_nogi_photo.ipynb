2. ダウンロードした画像のチェック<br>

そのためこのケースは削除の必要があるかも？

```python
img = Image.open('./data_bing/000002.jpg')
img = img.resize((100,100))
img
```

```python
folder = "./data_bing/"
max_num = len([f for f in os.listdir(folder) if os.path.isfile(os.path.join(folder, f)) and f.lower().endswith(('.jpg', '.jpeg', '.png'))])

record_num = 5
num_per_row = max_num // record_num + 1
num_per_col = max_num % record_num
fig, axes = plt.subplots(num_per_row, record_num, figsize=(10, 20))
axes_flat = axes.flatten()

for n in range(1, max_num + 1):
    if len(str(n)) == 1:
        num = f"00000{n}"
    else:
        num = f"0000{n}"

    img_path = f"{folder}{num}.jpg"
    try:
        img = mpimg.imread(img_path)
        axes_flat[n-1].imshow(img)
    except FileNotFoundError:
        axes_flat[n-1].text(0.5, 0.5, "missing", ha="center", va="center")
    axes_flat[n-1].axis('off')

plt.show()
```
