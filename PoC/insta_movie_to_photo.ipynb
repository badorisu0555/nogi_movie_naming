```python
def download_instagram_video(url, output_path="movie"):
    ydl_opts = {
        'outtmpl': f'{output_path}/nogi_movie_007.mp4',
        'quiet': False,
        'no_warnings': False,
    }
    
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        try:
            ydl.download([url])
            print("ダウンロード完了")
        except Exception as e:
            print(f"エラーが発生しました: {e}")

# 使用例
movie = movie_list[4]
download_instagram_video(movie)
```
